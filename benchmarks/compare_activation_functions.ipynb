# Compare custom implementation of activation functions with pytorch's

This notebook will show comparison between how the custom implementation of activation function fairs against torch's implementation

```python
import torch
from pytorch_fundamentals.activations.sigmoid import SigmoidActivationCustom
from pytorch_fundamentals.activations.tanh import TanhActivationCustom
```

## Reusable pieces of code

code written in the very next cell will be used through out the notebook

```python
# Tensor to be used throughout for comparing custom activation implementations with that of pytorch
X = torch.arange(-100, 100, 0.1)

# Function to be used to compare custom implementations with that of pytorch
def compare_implementation(X: torch.Tensor, custom_implementation_class: torch.nn.Module, pytorch_class: torch.nn.Module) -> None:
    """
    Function that generates a comparision between custom implementation of activation functions with that of pytorch'set

    :param X: Input tensor to base the comparison on
    :param custom_implementationc_class: Custom class that implements the activation function
    :param pytorch_class: Pytorch's implementation

    :returns: None
    """

    # Instantiate custom and pytorch activation classes
    custom_instance: torch.nn.Module = custom_implementation_class()
    pytorch_instance: torch.nn.Module = pytorch_class()

    # Generate the output after forward propagation of X through the loss functions
    output_custom: torch.Tensor = custom_instance.forward(X)
    output_pytorch: torch.Tensor = pytorch_instance.forward(X)

    # Compare the loss function outputs and provide the answer
    if torch.all(torch.isclose(output_custom, output_pytorch)):
        print("The implementation is correct")
    else:
        print("The implementation is incorrect")
```

## Sigmoid Non Linear Activation Function

```python
compare_implementation(X, custom_implementation_class=SigmoidActivationCustom, pytorch_class=torch.nn.Sigmoid)
```

## Hyperbolic Tangent (Tanh) Non Linear Activation Function

```python
compare_implementation(X, custom_implementation_class=TanhActivationCustom, pytorch_class=torch.nn.Tanh)
```
